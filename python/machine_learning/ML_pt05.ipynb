```python
import pandas as pd
#pipeline 이란 "스케일링, 학습 전처리 과정을 한번에 해주는 함수" 라고 암기해두기
```

```python
red_wine = pd.read_csv("../cv_data/winequality-red.csv", sep=";")
white_wine = pd.read_csv("../cv_data/winequality-white.csv", sep=";")

red_wine["color"] = 1.
white_wine["color"] = 0.
wine = pd.concat([red_wine, white_wine])

wine
```

```python

X = wine.drop(["color"], axis=1)
y = wine["color"]

X
```

```python
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler

estimators = [("scaler", StandardScaler()),
              ("clf", DecisionTreeClassifier())]

pipe = Pipeline(estimators)
```

```python
pipe.steps
```

```python
pipe.steps[0]
```

```python
pipe.steps[1]
```

```python
pipe["scaler"]
```

```python
pipe["clf"]
```

```python
pipe.set_params(clf__max_depth=2)
pipe.set_params(clf__random_state=13)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, 
                                                    random_state = 13,
                                                    stratify=y)
pipe.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred_tr = pipe.predict(X_train)
y_pred_test = pipe.predict(X_test)

#(accuracy_score(X_train, y_pred_tr))
#accuracy_score(X_test, y_pred_test)
```

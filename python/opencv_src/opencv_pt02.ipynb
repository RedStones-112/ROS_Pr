```python
import numpy as np
import cv2
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
```

```python

plt.imshow(img)
```

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
img_gray = cv2.imread("../cv_data/candies.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray, cmap="gray")
```

```python
new_img = cv2.resize(fix_img, (800,800)) 
# 크기변경 (주의 이미지가 변형될 수 있음)
plt.imshow(new_img)
```

```python
plt.imshow(cv2.flip(new_img, 0)) # 상하 반전
```

```python
plt.imshow(cv2.flip(new_img, 1)) #좌우 반전
```

```python
plt.imshow(cv2.flip(new_img, -1)) # 둘다 반전
```

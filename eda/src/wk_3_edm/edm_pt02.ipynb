```python
real_gu = {"종로구" : ["종로","혜화"],
           "중구" : ["남대문", "중부"],
           "용산구" : ["용산"],
           "성동구" : ["성동"],
           "광진구" : ["광진"],
           "동대문구" : ["동대문"],
           "중랑구" : ["중랑"],
           "성북구" : ["성북","종암"],
           "강북구" : ["강북"],
           "도봉구" : ["도봉"],
           "노원구" : ["노원"],
           "은평구" : ["서부","은평"],
           "서대문구" : ["서대문"],
           "마포구" : ["마포"],
           "양천구" : ["양천"],
           "강서구" : ["강서"],
           "구로구" : ["구로"],
           "금천구" : ["금천"],
           "영동포구" : ["영등포"],
           "동작구" : ["동작"],
           "관악구" : ["관악"],
           "서초구" : ["서초","방배"],
           "강남구" : ["강남","수서"],
           "송파구" : ["송파"],
           "강동구" : ["강동"]
           }
```

```python
real_gu.items()
```

```python
tmp_gu = []
for station in crime_station.index:
    for gu_name, station_name in real_gu.items():
        if station in station_name:
            tmp_gu.append(gu_name)
tmp_gu
    
```

```python
print(len(tmp_gu))
```

```python
crime_station.head()
```

```python
crime_station["구"] = tmp_gu
crime_station.head()
```

```python
crime_gu = crime_station.pivot_table(index="구", aggfunc=np.sum) #같은 인덱스끼리는 병합
crime_gu
```

```python
arrest_col = ["강도검거", "살인검거", "성범죄검거", "절도검거", "폭력검거"]
accur_col = ["강도발생", "살인발생", "성범죄발생", "절도발생", "폭력발생"]
```

```python
crime_gu[arrest_col].div(crime_gu[accur_col].values) * 100
```

```python
arrest_rate = ["성범죄검거율", "강도검거율", "살인검거율", "절도검거율", "폭력검거율"]
crime_gu[arrest_rate] = crime_gu[arrest_col].div(crime_gu[accur_col].values) * 100
crime_gu
```

```python
crime_gu[crime_gu[arrest_rate] > 100] = 100
crime_gu
```

```python
crime_gu.drop(columns=arrest_col, axis=1, inplace=True)
crime_gu
```

```python
crime_gu.rename(columns={"강도발생" : "강도",
                         "살인발생" : "살인",
                         "성범죄발생" : "성범죄",
                         "절도발생" : "절도",
                         "폭력발생" : "폭력"}, inplace=True)
crime_gu
```

```python
occur_col = ["강도", "살인", "성범죄", "절도", "폭력"]
```

```python
print(crime_gu.loc["강서구","절도"])
print(crime_gu["절도"].max())
```

```python
print(crime_gu.loc["강서구","절도"])
print(crime_gu.loc["강서구","절도"] / crime_gu["절도"].max())
```

```python
crime_gu_norm = crime_gu[occur_col] / crime_gu[occur_col].max()
```

```python
crime_gu_norm
```

```python
crime_gu_norm[arrest_rate] = crime_gu[arrest_rate]
crime_gu_norm.head()
```

```python
crime_gu_norm[occur_col].loc["강남구"].sum() / 5
```

```python
crime_gu_norm["범죄"] = np.mean(crime_gu_norm[occur_col], axis=1)
crime_gu_norm["검거"] = np.mean(crime_gu_norm[arrest_rate], axis=1)
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib
import seaborn as sns
```

```python
crime_gu_norm_sort = crime_gu_norm.sort_values(by="검거", ascending=False)
crime_gu_norm_sort.head()
```

```python
target_col = ["성범죄검거율", "강도검거율", "살인검거율", "절도검거율", "폭력검거율", "검거"]

plt.figure(figsize=(10,10))
sns.heatmap(crime_gu_norm_sort[target_col], annot=True, fmt="f", linewidths=.5, cmap="RdPu")
plt.title("범죄 검거 비율")
plt.show()
```

```python
target_col = ["성범죄", "강도", "살인", "절도", "폭력", "범죄"]
crime_gu_norm_sort = crime_gu_norm.sort_values(by="범죄", ascending=False)


plt.figure(figsize=(10,10))
sns.heatmap(crime_gu_norm_sort[target_col], annot=True, fmt="f", linewidths=.5, cmap="RdPu")
plt.title("범죄 발생 비율")
plt.show()
```

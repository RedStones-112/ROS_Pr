```python
fp1 = np.polyfit(data_result["인구수"], data_result["총계"], 1)
fp1
f1 = np.poly1d(fp1)
f1(400000)
fx = np.linspace(100000, 700000, 100) #시작값 끝값 지정된 개수의 리스트 생성
```

```python
def drawGraph4():
    plt.figure(figsize=(14,10))
    plt.scatter(data_result["인구수"], data_result["총계"], s = 50)
    plt.plot(fx, f1(fx), ls="dashed", lw=3, color="g")
    plt.xlabel("인구수")
    plt.ylabel("CCTV")
    plt.grid()
    plt.show()
drawGraph4()
```

```python
fp1 = np.polyfit(data_result["인구수"], data_result["총계"], 1)
#fp1
f1 = np.poly1d(fp1)
f1(400000)
fx = np.linspace(100000, 700000, 100)

data_result["오차"] = data_result["총계"] - f1(data_result["인구수"])

df_descendiong = data_result.sort_values(by= "오차", ascending=False)
df_ascendiong = data_result.sort_values(by= "오차", ascending=True)
```

```python
from matplotlib.colors import ListedColormap
color_step = ["#e74c3c", "#2ecc71", "#95a5a6", "#2ecc71", "#3498db", "#3498db"]
my_colormap = ListedColormap(color_step)
```

```python
def drawGraph5():
    plt.figure(figsize=(14,10))
    plt.scatter(data_result["인구수"], data_result["총계"],c=data_result["오차"],
                 s = 50, cmap=my_colormap)

    plt.plot(fx, f1(fx), ls="dashed", lw=3, color="gray")
    for n in range(5):
        plt.text(df_descendiong["인구수"][n] * 1.02,
                 df_descendiong["총계"][n] * 0.98,
                 df_descendiong.index[n], fontsize=15)
        
        plt.text(df_ascendiong["인구수"][n] * 1.02,
                 df_ascendiong["총계"][n] * 0.98,
                 df_ascendiong.index[n], fontsize=15)



    plt.xlabel("인구수")
    plt.ylabel("CCTV")
    plt.colorbar()
    plt.grid()
    plt.show()
drawGraph5()
```

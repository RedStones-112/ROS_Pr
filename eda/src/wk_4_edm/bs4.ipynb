```python
import re
import pandas as pd
import time
from urllib.request import urlopen, Request
from user_agent import generate_user_agent

url_sandwichs = "https://www.chicagomag.com/chicago-magazine/november-2012/best-sandwiches-chicago/"

req = Request(url=url_sandwichs, headers={"User-Agent":generate_user_agent()})

page = urlopen(req)
soup = BeautifulSoup(page, "html.parser")
```

```python
def get_info(url_info):
    req = Request(url=url_info, headers={"User-Agent":generate_user_agent()})
    page = urlopen(req)
    soup = BeautifulSoup(page, "html.parser")
    text = soup.find(class_="addy").get_text()
    text = re.split(".,",text)[0]
    price = re.search("\$\d+.(\d+)?",text).group()
    address = text[len(price)+1 :]
    #print(address)
    return price,address
    
    
    
```

```python
from urllib.parse import urljoin
f_url = "https://www.chicagomag.com"
list_sandwich = soup.find_all("div", "sammy")
sandwich_info = [[],[],[],[],[],[]]
for item in list_sandwich:
    url_sandwich = item.find("a")["href"]
    price, address = get_info(urljoin(f_url,url_sandwich))
    sandwich_info[0].append(item.find(class_="sammyRank").string) 
    sandwich_info[1].append(item.find("b").string)
    sandwich_info[2].append(re.split(("\n"), item.find(class_="sammyListing").get_text())[1])
    sandwich_info[3].append(url_sandwich)
    sandwich_info[4].append(price)
    sandwich_info[5].append(address)
```

```python
data = {"rank":sandwich_info[0],
        "menu":sandwich_info[1],
        "cafe":sandwich_info[2],
        "url":sandwich_info[3],
        "price":sandwich_info[4],
        "address":sandwich_info[5]}
df = pd.DataFrame(data)
df.to_csv("../../data/chicago_sandwich.csv",sep=",",encoding="utf-8")
```
